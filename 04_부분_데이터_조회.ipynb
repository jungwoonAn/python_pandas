### 1. 단일 열 조회
    데이터프레임.열이름
    데이터프레임["열이름"]
- 열 이름에. 또는 공백 등이 포함되어 있을 경우에는 [ ]형식을 사용해야 함

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.sepal_length
```

### 2. loc를 이용한 이름으로 조회
- **loc[ 행이름, 열이름]** 형식으로 행/열의 이름으로 부분 데이터 조회, 열 이름 생략 가능

```python
iris.loc[0]  # 행의 이름이 0인 행
```

```python
iris.loc[:, 'sepal_length']  # 열 이름이 'sepal_length'인 모든 행
```

```python
# 행 이름 또는 열 이름을 잘 못 지정하면 에러 발생함
iris.loc['sepal_length']
```

- **loc[ 행이름시작:행이름끝, 열이름시작:열이름끝]** 형식으로 부분 데이터 조회 가능
- loc는 이름끝(from:to 형식에서 to 이름)을 포함

```python
iris.loc[0:5]  # 0부터 5까지(to 포함), 모든 열 출력
```

```python
iris.loc[:, 'sepal_width':'petal_width']
```

### 3. iloc를 이용한 인덱스로 조회
     iloc[행인덱스, 열인덱스]
     iloc[행시작인덱스:행끝인덱스, 열시작인덱스:열끝인덱스]
- iloc는 끝인덱스(from:to 형식에서 to 인덱스)는 포함 안함

```python
iris.iloc[0:5]  # iris.loc[0:5]의 결과와 다름, iloc는 to 인덱스 포함 안함
```

```python
iris.iloc[0:5, 0:3]
```

    iloc[from:to, from:to, by=1]
-  iloc는 끝인덱스(from:to 형식에서 to 인덱스)는 포함 안함

```python
iris.iloc[0:10:2, ::2]  # 0행부터 10행까지 매2번째 행 출력, 첫열부터 마지막열까지 매2번째 열 출력
```

### 4. 조건으로 조회하기
- **loc[행조건]** : 행 조건에 맞는 모든 열을 반환

```python
iris.loc[iris.species=='versicolor'].head()
```

-  **loc[행조건, 열리스트]** : 행 조건에 맞는 지정한 열을반환

```python
iris.loc[iris.species=='versicolor',
        ['sepal_length','species']].head()
```

- **loc[& 또는 |를 이용한 중복조건]**

```python
iris.loc[(iris.species=='versicolor') &
         (iris.sepal_length > 6.5)].head()
```

[정리]  

    - loc[ ] : 이름으로 조회, 조건으로 조회 가능
    - iloc[ ] : 번호로 조회
