# 03_이름 지정하기
### 1. 열 이름 지정하기
- columns 속성을 이용해서 열 이름을 지정하거나 조회

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.columns  # columns 속성은 열 이름 정보를 가지고 있음
```

```python
iris.columns = ['sl','sw','pl','pw','sp']  # columns 속성에 리스트 형식으로 열의 이름들을 지정할 수 있음
```

```python
iris.head()
```

### 2. 행 이름 지정하기
- index 속성을 이용해서 행 이름을 지정하거나 조회

```python
iris.index
```

```python
iris.index = range(150,300)  # index 속성 이용해서 행 이름지정
```

```python
iris.head()
```

### 3. 레벨 이름 지정하기
- 행 또는 열은 2개 이상의 레벨 이름을 가질 수 있음

##### 1) 열에 2개 이상의 레벨 이름 지정하기

```python
iris.columns = \
    [['sepal','sepal','petal','petal','species'],
     ['length','width','length','width','species']]
```

```python
iris.head()
```

```python
iris.columns.names = ['sps', 'lw']
```

```python
iris.head()
```

```python
# 레벨별 열 이름을 지정해 그룹된 열을 쉽게 조회할 수 있음
iris['sepal'].head()
```

```python
iris['petal'].head()
```

##### 2) 행에 2개 이상의 레벨 이름 지정하기

```python
iris.index = [['setosa' for i in range(50)] +
              ['versicolor' for i in range(50)] +
              ['virginica' for i in range(50)],
              list(range(150))]
```

```python
iris.head(10)
```

```python
iris.index.names = ['sp','rownum']
```

```python
iris.head()
```
