### 1. dropna
- **dropna()** : **결측치(누락된 값, NA, NaN)를 포함한 행 또는 열을 제거**해줌
        DataFrame.dropna(*, axis=0, how=<no_default>, thresh=<no_default>,
        subset=None, inplace=False, ignore_index=False)
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html

```python
iris_x.dropna().head()
```

```python
iris_x.dropna(thresh=2).head(10)  # thresh=2 : NaN값이 아닌 값이 2개 이상이면 삭제 안함
```

```python
iris_x.dropna(subset=['sepal_length', 'sepal_width']).head()  # subset 누락된 값을 찾을 열을 지정하여 제거
```

```python
iris_x.dropna(inplace=True)
```

```python
iris_x.head()
```

### 2. fillna
-  **fillna()** : 결측치(NA, NaN)를 주어진 방법으로 채움
        DataFrame.fillna(value=None, *, method=None, axis=None, inplace=False,
        limit=None, downcast=<no_default>)
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html

   -  **method** : {'backfill', 'bfill', 'pad', 'ffill', None}, 기본값 None,  
   재 색인화된 채우기에 사용될 방법을 지정.
    - **'ffill'**: 이전의 유효한 관측값을 이용해 채움
    - **'bfill'**: 이후의 유효한 관측값을 사용해서 채움

```python
iris = sns.load_dataset('iris')
iris_x = iris.iloc[:, :-1]

import random
random.seed(1)

for col in range(4):
    iris_x.iloc[[random.sample(range(len(iris)), 20)], col] = float('nan')
```

```python
iris_x.fillna(value=0).head()  # NaN -> 0으로
```

```python
iris_x.fillna(method='ffill').head(10)  # 이전에 유효한 값으로 채워짐
```

```python
iris_x.head(10)
```

- 평균값으로 채우기

```python
import numpy as np
```

```python
np.round(iris_x.mean(), 1)
```

```python
iris_x.fillna(value=np.round(iris_x.mean(), 1)).head()
```

- limit 속성 : NaN값을 채울 갯수 지정

```python
iris_x.fillna(value=np.round(iris_x.mean(), 1),
              limit=2).head(10)
```

```python
iris_x.head(10)
```

### 3. replace
-  **replace()** : to_replace속성에 주어진 값을 value 값으로 바꿈
- 대체될 값을 찾는 방법을 숫자(numeric), 문자(str) 또는 정규식(regex)으로 지정할 수 있음
        DataFrame.replace(to_replace=None, value=<no_default>, *,
        inplace=False, limit=None, regex=False, method=<no_default>)
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html

- 상수값으로 대체

```python
iris_x.replace(float('nan'), 10).head(10)
```

- 채울 방법 지정

```python
iris_x.sepal_length.replace([5, 4.6], method='bfill').head(10)
```

- 정규표현식 이용

```python
iris.replace(r"^se[a-z]*", "set", regex=True).head()
```

```python
iris.head()
```

```python
iris.replace(regex=r"^se[a-z]*", value="set").head()
```

- 정규표현식을 이용한 replace

```python
df = pd.DataFrame({'A': ['bat', 'foo', 'bait'],
                   'B': ['abc', 'bar', 'xyz']})
```

```python
df
```

```python
df.replace(r"^ba.$", 'new', regex=True)
```

```python
df.replace({'A': r"^ba.$"}, {'A': 'new'}, regex=True)
```

```python
df.replace(regex=r"^ba.$", value='new')
```

```python
df.replace(regex={r"^ba.$": 'new', 'foo': 'xyz'})
```

```python
df.replace(regex=[r"^ba.$", "foo"], value='new')
```

- Series 데이터를 이용한 replace

```python
s
```

```python
s.replace({'a': None})
```

```python
s.replace('a', None)
```

```python
s.replace(to_replace='a', value=None, method='pad')
```

### 4. where
- **where()** : 하나 이상의 조건에 대한 데이터프레임을 확인하고 그에 따라 결과를 반환  
        DataFrame.where(cond, other=nan, *, inplace=False, axis=None, level=None)
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.where.html

    -  **cond** : boolean NDFrame, array-like, 호출가능객체(callable), 만일
cond가 True이면 원래 값을 유지  
    - **other** : scalar, NDFrame, callable, cond가 False인 항목은 other의 해당
값으로 대체
    - if/elif를 대체할 수 있음

```python
iris = sns.load_dataset('iris')
iris_x = iris.iloc[:, :-1]
```

```python
iris_x.where(iris_x > 5).head(10)
```

```python
iris_x.where(iris_x > 5, other=0).head(10)
```

```python
df = pd.DataFrame(np.arange(10).reshape(-1, 2),  # reshape(-1, 2) : 열은 2열, 행(-1)은 알아서 지정
                  columns=['A','B'])
```

```python
df
```

```python
df.where(df%3 == 0, -df)  # df.where(df%3 == 0, other=-df)
```

- 넘파이 where()함수와 사용법이 조금 다름

```python
np.where(df%3 == 0, df, -df)  # np.where(조건식, 참, 거짓)
```

- mask()함수는 where()함수와 조건이 반대

```python
df.mask(~(df%3 == 0), -df)
```

### 5. astype
- **astype()** : 판다스의 객체(데이터프레임과 시리즈)의 모든 요소의 **자료형을 바꿈**
        DataFrame.astype(dtype, copy=None, errors='raise')
    -  **dtype** : 자료형 또는 딕셔너리, numpy.dtype 또는 파이썬의 자료형을 사용하여 전체 판다스 객체를 같은 유형으로 형변환 할 수 있음.  
    {col: dtype, …}처럼 딕셔너리 형식을 사용했을 경우 col은 열의 이름이고, 형변환하는 numpy.dtype 또는 파이썬의 자료형

```python
iris_x.astype(int).head()
```

```python
iris_x.astype({'sepal_length': int, 'sepal_width': int}).head()
```

    - replace() : 값을 일괄 변경, 조건, 리스트, 정규식을 이용해서 변경
