### 2. 분산, 표준편차
-  **var()** : 분산(variance)
        DataFrame.var(axis=0, skipna=True, ddof=1, numeric_only=False, **kwargs)
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.var.html
- **std()** : 표준편차(standard deviation)
        DataFrame.std(axis=0, skipna=True, ddof=1, numeric_only=False, **kwargs)
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.std.html

```python
iris.select_dtypes(include='number').var()
```

```python
iris.select_dtypes(include='number').std()
```

- ddof(델타 자유도, Delta Degree of Freedom) 매개변수의 기본값이 1  
    - ddof=1 : 표본분산과 표본표준편차를 계산함  
    - ddof=0 : 모분산, 모표준편차를 계산함
- 넘파이 ddof 기본값은 0, 판다스의 ddof 기본값은 1

```python
iris.select_dtypes(include='number').std(ddof=1)  # 표본표준편차
```

```python
iris.select_dtypes(include='number').std(ddof=0)  # 모표준편차
```

### 3. 공분산, 상관계수
- **cov()** : 각 열들의 공분산(covariance) 쌍을 계산  
  공분산은 데이터 크기에 따라서 값이 다르게 나타날 수 있음.
        DataFrame.cov(min_periods=None, ddof=1, numeric_only=False)
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.cov.html
- **corr()** : 각 열들의 상관계수(correlation coefficient) 쌍을 계산  
  공분산을 -1 ~ 1로 표준화 시킨 것.
        DataFrame.corr(method='pearson', min_periods=1, numeric_only=False)
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html

-  **method** : 상관계수를 계산할 방법을 지정함  
  “pearson”, “kendall”, “spearman” 중 하나를 지정할 수 있음

```python
iris.select_dtypes(include='number').cov()
```

```python
iris.select_dtypes(include='number').corr()  # 상관관계는 숫자로 보는것보다 그림으로 보는 경우가 많음
```

### 4. 요약 통계량
- **describe()** : 요약 통계량 정보를 출력해 줌
        DataFrame.describe(percentiles=None, include=None, exclude=None)
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.describe.html

    - **percentiles** : 출력에 포함될 백분위 수를 0~1사이의 값으로 지정.  
      기본 값은 [.25, .5, .75]. 25%, 50%, 75% 위치 데이터를 출력  
    - **include** : 출력에 포함될 데이터의 유형을 지정합니다.  
      None(기본값) 이면 모든 숫자 타입 열들을 출력에 포함시킴.  
      “all”이면 모든 열을 포함.  
      정수형이면“int64”, 논리형이면 “bool”, 실수형이면 “float64” 등으로 지정함.  
    - **exclude** : 출력에서 제외할 데이터의 유형을지정.  
      None(기본값) 이면 아무것도 제외시키지 않음  

---


- 숫자 데이터
    - 결과의 인덱스에는 **count, mean, std, min, max** 및 **하위 백분위 수, 상위 백분위 수 및 상위 백분율**이 포함  
    - 기본적으로 하위 백분위 수는 25이고 상위 백분위 수는 75입니다. 50 백분위 수는 중앙값과 같음  
- 객체 데이터(예: 문자열 또는 타임스탬프)  
    - 결과색 인에 **count, unique, top** 그리고 **freq**가 포함
    - top가 가장 일반적인 값.  
    - 여러 오브젝트 값이 가장 높은 count를 갖는 경우, count와 top 결과는 가장 높은 count를 갖는 오브젝트 값 중에서 임의로 선택.  
- DataFrame을 통해 제공되는 혼합 데이터 유형
    - **기본값은 숫자 열의 분석만 반환**  
    - 데이터프레임이 **숫자 열이 없는 개체 및 범주 데이터로만 구성된 경우** 기본값은 **개체 열과 범주형 열 모두의 분석을 반환**  
    - include='all' 매개변수가 제공되면 결과에는 각 유형의 속성이 결합됨

### 5. 기본 요약 통계량  
- 기본적으로 숫자 데이터의 요약 통계량이 출력  
- include=‘all’을 추가하면 모든 열의 통계량을 볼 수 있음

```python
iris.describe()
```

```python
iris.species.describe()  # 데이터프레임이 숫자열이 없는 경우 -> 개채열과 범주형 열 모두의 분석을 반환
```

### 6. include와 exclude
-  include 및 exclude를 사용하여 요약 통계량에 포함될 열의 타입을 포함하거나 제외함  
   include=‘all’일 경우 모든 열 포함

```python
iris.describe(include='all')
```
